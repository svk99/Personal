```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,BaggingClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score,accuracy_score
import math
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv('/content/Churn Modeling.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['Exited']=df['Exited'].astype('category')
```

```python
df.describe()
```

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
dff=pd.get_dummies(df,columns=['Geography','Gender'])
```

```python
dff
```

```python
feature=dff.drop(['Exited'],axis=1)
target=dff['Exited']
x_train,x_test,y_train,y_test = train_test_split(feature,target,test_size=0.2,random_state=42)
model=[LogisticRegression(),DecisionTreeClassifier(),RandomForestClassifier(),AdaBoostClassifier(),BaggingClassifier(),SVC()]
for m in model:
  print('Model:',m)
  m.fit(x_train,y_train)
  test_pred=m.predict(x_test)
  print("Accuracy: ",accuracy_score(test_pred,y_test))
```
